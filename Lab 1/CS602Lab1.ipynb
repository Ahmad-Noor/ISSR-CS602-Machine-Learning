## Implementing linear regression with one variable to predict profits for a food truck.The file data.csv contains the dataset for our linear regression problem. The first column is the population of a city and the second column is the profit of a food truck in that city.
## Now we have to predict profit for given population city

```python
fileurl = 'https://raw.githubusercontent.com/Salma-Alhagry/CS602/master/Linear%20Regression/data.csv'
data = pd.read_csv(fileurl)
```

```python
x = data['population'] #feature
y = data['profit'] #predictor
```

```python
plt.scatter(x,y)
plt.xlabel("popultion")
plt.ylabel("profit")
plt.show()
```

```python
x = np.array(x)
y = np.array(y)
```

```python
x = x.reshape(len(x), 1)
y = y.reshape(len(y), 1)
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test= train_test_split(x, y, test_size= 0.4)
```

```python
from sklearn.linear_model import SGDRegressor
```

```python
sgd = SGDRegressor(loss="squared_loss", max_iter=50)
```

```python
sgd.fit(X_train, y_train)
```

```python
print(sgd.intercept_)
print(sgd.coef_)
```

```python
predict_y = sgd.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
print(mean_squared_error(y_test, predict_y))
```

```python
plt.scatter(x,y)
plt.plot(x, sgd.predict(x), color="red")
plt.xlabel("popultion")
plt.ylabel("profit")
plt.show()
```
